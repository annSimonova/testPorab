```python
import matplotlib.pyplot as plt


def parab(x):
    return x * x


x = [i for i in range(-10, 11)]
y = [parab(i) for i in x]

plt.plot(x, y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("y = x^2")
plt.grid()
plt.show()
```

```python
N = 9
def const():
    return N

x = [i for i in range(-10, 11)]
y = [parab(i) for i in x]
z = [const() for i in x]

plt.plot(x, y)
plt.plot(x, z)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("y = x^2")
plt.grid()
plt.show()
```
